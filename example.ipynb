# Example of the bivariate GEV random vectors generator

The following example illustrates how to generate bivariate GEV random vectors using the MCMC algorithm provided in Theorem 3 of McFadden (1995). To recall, let the GEV distribution function be equal to:

$$F(\varepsilon_1,\varepsilon_2) = exp(-(e^{-\varepsilon_1/s} + e^{-\varepsilon_2/s})^s)$$

Where $s>0$ is a dissimilarity parameter that governs the correlation between $\varepsilon_1$ and $\varepsilon_2$. In the particular case of $s=1$, $\varepsilon_1$ and $\varepsilon_2$ are independent.

The reader can refer to McFadden (1995), Theorem 3 for the details about the MCMC algorithm. This example will limit to show how the GEV random number generator works.

## Step 1: Preliminary definitions

First, we load the required packages:

```python
# Load packages
import numpy as np
from MCMC_GEV_functions import gev_rand
import matplotlib.pyplot as plt
```

Then, we set the number of random draws, and the dissimilarity parameter. We will generate random numbers for $s=0.1$, $s=0.5$ and $s=1$:

```python
# Set Number of random draws, and dissimilarity parameter
N = 10000
s1 = 0.1
s2 = 0.5
s3 = 1
```

## Step 2: Random vector generation

Now, we will generate three sets of GEV random vectors, one for each value of $s$. The *burn_ratio* parameter first generates additional *burn_ratio %* draws, which are dropped in the end of the routine.

```python
dr1 = gev_rand(N,s1,burn_ratio=0.1)
dr2 = gev_rand(N,s2,burn_ratio=0.1)
dr3 = gev_rand(N,s3,burn_ratio=0.1)
```

## Step 3: Plots

We will generate a set of scatter plots with two histograms attached to their axes. Each scatter plot represents the joint distribution of $\varepsilon_1$ and $\varepsilon_2$, while each histogram represents an empirical marginal distribution of these random variables.

The following code generates a plot for *dr1*:

```python
# Range of the histograms' y-axis:
hist_y_range = (0,0.5)

# Limits of the scatterplot (and hence of the histograms' x-axis):
scatter_range = (-3,10)

# Generate the figure for the scatter plot
scatter_axes = plt.subplot2grid((3, 3), (1, 0), rowspan=2, colspan=2,xlim=scatter_range,ylim=scatter_range,xlabel=r'$\varepsilon_1$',ylabel=r'$\varepsilon_2$')

# Generate the figures for the histograms
x_hist_axes = plt.subplot2grid((3, 3), (0, 0), colspan=2,ylim=hist_y_range,sharex=scatter_axes)
x_hist_axes.tick_params(axis="x", labelbottom=False)

y_hist_axes = plt.subplot2grid((3, 3), (1, 2), rowspan=2,xlim=hist_y_range,sharey=scatter_axes)
y_hist_axes.tick_params(axis="y", labelleft=False)

# Put the data in the scatter plot and histograms
scatter_plot = scatter_axes.plot(dr1[:,0], dr1[:,1], '.')
hist_x = x_hist_axes.hist(dr1[:,0],density=True,bins = int(N*0.01))
hist_y = y_hist_axes.hist(dr1[:,1],density=True,bins = int(N*0.01), orientation='horizontal')

plt.tight_layout()
plt.show()
```
